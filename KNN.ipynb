```r
#User defined function for KNN classifier on Iris dataset.

iris
summary(iris)
dim(iris)
```

```r
#Splitting and sampling dataset in 7:3 ratio.
set.seed(420)
shuffled_data = iris[sample(1:nrow(iris)), ]
part <- sample(2, nrow(shuffled_data), replace = TRUE, prob = c(0.70, 0.30))

train<- iris[part == 1,]

test<- iris[part == 2,]

test
train
```

```r
#KNN Classifier

#Calculating the euclidean distance

ED<-function(data1,data2){
distance=0
  for (i in (1:(length(data1)-1))){
    distance=distance+(data1[i]-data2[i])^2
  }
  return(sqrt(distance))
}

#Finding maximum of 3 integers
maximum=function(a,b,c)
    {
	    max=a;
	    if(b>max)
		    max=b;
	    if(c>max)
		    max=c;
	    return(max);
    }

#Writing the function to predict kNN
knnPredictor <- function(test, train, kValue){
  predictedSpecies <- c()
  actualSpecies <- c()
  #LOOP-1
  for(i in c(1:nrow(test))){   
    distance = c()          
    species = c()
    setosa =0              
    versicolor = 0
    virginica = 0
    
    #LOOP-2-looping over train data 
    for(j in c(1:nrow(train))){
      
      distance<-append(distance, as.double(ED(test[i,], train[j,])))
      species<-append(species, as.character(train[j,][[5]]))
    }
    class.df <- data.frame(distance,species) 
    class.df<-class.df[order(distance), ]     #sorting dataframe
    class.df<- class.df[1:kValue,] 


     for(k in c(1:nrow(class.df))){
      if(as.character(class.df[k,][[2]]) == "setosa"){
        setosa = setosa + 1
      }else if(as.character(class.df[k,][[2]]) == "versicolor"){ #classifying the result
        versicolor = versicolor + 1
      }else
        virginica = virginica + 1
    } 

    actualSpecies <- append(actualSpecies,test[i,][[5]])

    maxResult = maximum(setosa,versicolor,virginica)
    if(maxResult == setosa)
       predictedSpecies <- append(predictedSpecies,"setosa")
    else if(maxResult == versicolor)
       predictedSpecies <- append(predictedSpecies,"versicolor") # Prediction of the output
    else
       predictedSpecies <- append(predictedSpecies,"virginica")        
  }
  finalResult <- data.frame(actualSpecies,predictedSpecies)
  return (finalResult)
} 

#Predicting the value for K=3
K=3
finalData = knnPredictor(test, train, K)
finalData

#Finding accuracy of the model
correctPredict = 0
sampleSize = nrow(finalData)

for(k in c(1:nrow(finalData)))
{
  if(finalData[k,][1] == finalData[k,][2])
     correctPredict = correctPredict + 1
}

accuracy <- correctPredict/sampleSize * 100
cat("\n")
print(paste("Accuracy of the model = ",accuracy))
```
